## Outlier Detection and remover

```python
sns.boxplot(df, orient="h", palette="Set2")
plt.show()
```

In this boxplot for whole dataset we can see the outliers for each columns.The outliers visible outside of the whiskers of the boxplot.

```python
def replace_outliers_with_mean(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = np.where((df[column] < lower_bound) | (df[column] > upper_bound),
                          df[column].mean(), df[column])
    return df

for column in df.select_dtypes(include=np.number).columns:
    df = replace_outliers_with_mean(df, column)

sns.boxplot(df, orient="h", palette="Set2")
plt.show()
df.shape
```

```python
plt.figure(figsize=(20,18))
sns.heatmap(df.corr(), annot=True,linewidths=.9, cmap="Purples")
```

```python
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
features_corr = df.corr()['diagnosis'].abs().sort_values(ascending=False)

# Display the top features with the highest correlation to diagnosis
print(features_corr.head(10))

# Visualization of top features' correlations with diagnosis
top_features = features_corr.index[1:10]
plt.figure(figsize=(10, 6))
sns.barplot(x=features_corr[1:10], y=top_features)
plt.title("Top Features Correlated with Diagnosis")
plt.xlabel("Correlation Coefficient")
plt.ylabel("Features")
plt.show()
```

In this plot you can see the top 10 features whis correlated with diagnosis.

*   concave points_worst is highly correlated with the diagnosis.
*   The correlation coefficient of diagnosis with radius_mean is low.



```python
# Group data by diagnosis and calculate descriptive statistics for numerical features
grouped_data = df.groupby('diagnosis').agg(['mean', 'std', 'min', 'max'])
print(grouped_data)

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean']
for i, feature in enumerate(features):
    row = i // 2
    col = i % 2
    ax = axes[row, col]

    sns.histplot(df[df['diagnosis'] == 0][feature], label='Benign', kde=True, ax=ax)
    sns.histplot(df[df['diagnosis'] == 1][feature], label='Malignant', kde=True, ax=ax)

    ax.set_title(f'Distribution of {feature} by Diagnosis')
    ax.legend()

plt.tight_layout()
plt.show()
```

* **Differences in average values:** Peaks of histograms in different positions for benign and malignant tumors indicate differences in the typical value of the feature.
* **Differences in data spread:** Wider histograms or longer tails for malignant tumors suggest more variability in the feature for those tumors.
* **Distribution shapes:**
  *   **Bell-shaped:** Feature is likely normally distributed.
  *   **Skewed:** Feature values are concentrated on one side.

  *   **Multiple peaks:** Potential subgroups within the data.
*   **Overlap:**
  *   **Significant overlap:** Feature may not be a strong predictor of diagnosis.
  *   **Little overlap:** Feature could be a good predictor.

* 1st plot : 'radius_mean' has a significantly higher mean and larger range for malignant tumors compared to benign tumors, it suggests that larger mean radius values are an indicator of malignancy.
* 2nd plot : 'texture_mean' has a significantly higher mean and lower range for malignant tumors copared to benign tumors, it suggests that larger mean texture values are a non-cancerous growth of cells that does not spread to other parts of the body.
* 3rd plot : we can see the 3rd plot is similar like the 1st plot i,e. the larger values of 'perimeter_mean' indicator of malignancy.
* 4th plot : In this plot we can see that larger values of 'area_mean' are a non-cancerous growth of cells that does not spread to other parts of the body.

```python
import statsmodels.api as sm
import pylab

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean']
for i, feature in enumerate(features):
    row = i // 2
    col = i % 2
    ax = axes[row, col]

    sm.qqplot(df[feature], line='45', ax=ax)
    ax.set_title(f'QQ-plot for {feature}')
plt.tight_layout()
plt.show()

# Skewness and Kurtosis:
from scipy.stats import skew, kurtosis
for feature in features:
    print(f"Skewness for {feature}: {skew(df[feature])}")
    print(f"Kurtosis for {feature}: {kurtosis(df[feature])}")
```

* radius_mean: Appears to be right-skewed for both benign and malignant tumors, but the malignant tumors have a longer tail.
* texture_mean: Might be slightly right-skewed for both, with less difference between the two groups.
* perimeter_mean: Similar to radius_mean, a right-skewed distribution, with a longer tail for malignant tumors.
* area_mean: Shows right skewness for both categories, with higher values for malignant cases.

```python
from scipy.stats import ttest_ind

# Perform t-tests for each feature
for feature in df.columns.drop('diagnosis'):
    group0 = df[df['diagnosis'] == 0][feature]
    group1 = df[df['diagnosis'] == 1][feature]
    t_statistic, p_value = ttest_ind(group0, group1)
    print(f"Feature: {feature}")
    print(f"  T-statistic: {t_statistic}")
    print(f"  P-value: {p_value}")
    if p_value < 0.05:  # Significance level of 0.05
        print(f"  Significant difference between groups (p < 0.05)")
    else:
        print(f"  No significant difference between groups")
    print("---")
```
