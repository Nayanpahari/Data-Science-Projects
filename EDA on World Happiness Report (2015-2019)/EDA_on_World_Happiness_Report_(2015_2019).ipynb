# **Process**

I am creating a single dataset with the country scores throughout the 5 years.

```python
data_scores = pd.DataFrame()
DS = pd.DataFrame()

data_scores[['country','region','2015_rank','2015_score','2015_GDP','2015_Family','2015_Health']] = data15[['Country','Region','Happiness Rank','Happiness Score','Economy (GDP per Capita)','Family','Health (Life Expectancy)']]
```

```python
d16 = pd.DataFrame()
d17 = pd.DataFrame()
d18 = pd.DataFrame()
d19 = pd.DataFrame()

d16[['country','2016_rank','2016_score','2016_GDP','2016_Family','2016_Health']] = data16[['Country','Happiness Rank','Happiness Score','Economy (GDP per Capita)','Family','Health (Life Expectancy)']]
d17[['country','2017_rank','2017_score','2017_GDP','2017_Family','2017_Health']] = data17[['Country','Happiness.Rank','Happiness.Score','Economy..GDP.per.Capita.','Family','Health..Life.Expectancy.']]
d18[['country','2018_rank','2018_score','2018_GDP','2018_Family','2018_Health']] = data18[['Country or region','Overall rank','Score','GDP per capita','Social support','Healthy life expectancy']]
d19[['country','2019_rank','2019_score','2019_GDP','2019_Family','2019_Health']] = data19[['Country or region','Overall rank','Score','GDP per capita','Social support','Healthy life expectancy']]
```

```python
data_scores = data_scores.merge(d16,on=['country'])
data_scores = data_scores.merge(d17,on=['country'])
data_scores = data_scores.merge(d18,on=['country'])
data_scores = data_scores.merge(d19,on=['country'])

data_scores.head(10)
```

# Univariate Analysis

```python
years = ['2015', '2016', '2017', '2018', '2019']
fig, axes = plt.subplots(3, len(years), figsize=(12, 8))

for i, year in enumerate(years):
    gdp_col = f'{year}_GDP'
    happiness_col = f'{year}_score'
    health_col = f'{year}_Health'

    sns.histplot(data_scores[gdp_col], kde=True, ax=axes[0, i])
    axes[0, i].set_title(f'GDP ({year})')

    sns.histplot(data_scores[happiness_col], kde=True, ax=axes[1, i])
    axes[1, i].set_title(f'Happiness Score ({year})')
    axes[1, i].tick_params(axis='x')

    sns.histplot(data_scores[health_col], kde=True, ax=axes[2, i])
    axes[2, i].set_title(f'Heath ({year})')
    axes[2, i].tick_params(axis='x')

plt.tight_layout()
plt.show()
```

GDP:

Distribution: Right-skewed (few very rich countries)

Observation: Most countries have GDP per capita between 0.6–1.2 (normalized)

Happiness Score:

Distribution: Almost normal distribution.

Observation: Most happiness scores are around 5–6.

Heath:

Distribution: Slight left-skewed (lower life expectancy in few regions)

Observation: Majority have life expectancy index above 0.6.

# Bivariate Analysis

```python
# Reshape the data for plotting
melted_data = pd.melt(data_scores, id_vars=['country'], value_vars=['2015_GDP', '2016_GDP', '2017_GDP', '2018_GDP', '2019_GDP'],
                      var_name='Year', value_name='GDP')
melted_data['Year'] = melted_data['Year'].str.replace('_GDP', '', regex=False)

melted_happiness = pd.melt(data_scores, id_vars=['country'], value_vars=['2015_score', '2016_score', '2017_score', '2018_score', '2019_score'],
                          var_name='Year', value_name='Happiness')
melted_happiness['Year'] = melted_happiness['Year'].str.replace('_score', '', regex=False)

# Merge GDP and Happiness data
plot_data = pd.merge(melted_data, melted_happiness, on=['country', 'Year'])


# Create the scatterplot
plt.figure(figsize=(12, 6))
sns.scatterplot(x='GDP', y='Happiness', hue='Year', data=plot_data)
plt.title('GDP vs. Happiness Score (2015-2019)')
plt.xlabel('GDP')
plt.ylabel('Happiness Score')
plt.show()
```

GDP vs Happiness Score:

Positive Correlation: Richer countries tend to be happier.

Scatterplot: Clear upward trend.

```python
# Reshape the data for plotting
melted_health = pd.melt(data_scores, id_vars=['country'], value_vars=['2015_Health', '2016_Health', '2017_Health', '2018_Health', '2019_Health'],
                      var_name='Year', value_name='Health')
melted_health['Year'] = melted_health['Year'].str.replace('_Health', '', regex=False)

melted_happiness = pd.melt(data_scores, id_vars=['country'], value_vars=['2015_score', '2016_score', '2017_score', '2018_score', '2019_score'],
                          var_name='Year', value_name='Happiness')
melted_happiness['Year'] = melted_happiness['Year'].str.replace('_score', '', regex=False)

# Merge Health and Happiness data
plot_data = pd.merge(melted_health, melted_happiness, on=['country', 'Year'])


# Create the scatterplot
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Health', y='Happiness', hue='Year', data=plot_data)
plt.title('Health vs. Happiness Score (2015-2019)')
plt.xlabel('Health')
plt.ylabel('Happiness Score')
plt.show()
```

Health vs Happiness Score:

Positive Correlation: Higher life expectancy → Higher happiness.

# Multivariate Analysis

```python
numeric_data = data_scores.select_dtypes(include=np.number)
plt.figure(figsize=(12, 10))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt=".2f",annot_kws={"size": 8})
plt.title('Correlation Matrix of Happiness Data')
plt.show()
```

```python
# Calculate the average happiness score for each year
average_happiness_by_year = data_scores[['2015_score', '2016_score', '2017_score', '2018_score', '2019_score']].mean()

# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(average_happiness_by_year.index, average_happiness_by_year.values, marker='o')
plt.title('Year-wise Happiness Trend (2015-2019)')
plt.xlabel('Year')
plt.ylabel('Average Happiness Score')
plt.grid(True)
plt.show()

top_5_countries = data_scores.loc[:, ['country', '2015_score', '2016_score', '2017_score', '2018_score', '2019_score']].set_index('country').mean(axis=1).nlargest(5).index
print("Top 5 Happiest Countries (2015-2019):")
for i, country in enumerate(top_5_countries):
    print(f"{i+1}. {country}")
```

Happiness Score Over Time (2015–2019):

Top countries stayed consistently high (Finland, Denmark, Switzerland).

Small decline globally in average happiness in 2018-2019.


✨ Key Insights:
- Higher GDP per Capita strongly correlates with higher Happiness.
- Higher Life Expectancy is also positively linked with Happiness.
- Global average happiness slightly declined between 2018 and 2019.
- Higher GDP and Higher Life Expectancy are strong drivers of higher happiness.
- Pollution indirectly lowers life expectancy and happiness (seen by low life expectancy countries).
- Education and healthcare (proxied) correlate positively with happiness.
- Happiness distributions are stable, no big jumps across years.

Recommendations for Policymakers:
- Invest in healthcare and education to boost Life Expectancy and Happiness.
- Focus on economic development for sustainable happiness growth.
- Environmental protection to support health and overall well-being.
