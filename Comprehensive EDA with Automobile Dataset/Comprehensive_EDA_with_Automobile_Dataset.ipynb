# **The column renames based on the following :**

**symboling:** Represents the insurance risk rating of a car, with higher values indicating greater risk.

**normalized_losses:** Represents the average loss payment per insured vehicle year. This is a normalized value, allowing for comparison across different car models.

**make:** Identifies the manufacturer of the car (e.g., Toyota, Honda, Ford).

**fuel_type:** Specifies the type of fuel used by the car (e.g., gas, diesel).

**aspiration:** Describes the engine aspiration type (e.g., standard, turbo).

**num_doors**: Indicates the number of doors on the car (e.g., two, four).

**body_style:** Specifies the type of car body (e.g., sedan, hatchback, wagon).

**drive_wheels:** Describes the type of drive system (e.g., front-wheel drive, rear-wheel drive, four-wheel drive).

**engine_location:** Indicates the location of the engine (e.g., front, rear).

**wheel_base:** Represents the distance between the front and rear wheels.

**length, width, height:** Specify the dimensions of the car.

**curb_weight:** Represents the weight of the car without passengers or cargo.

**engine_type:** Describes the type of engine (e.g., dohc, ohc, ohcv).

**num_cylinders:** Indicates the number of cylinders in the engine.

**engine_size:** Represents the size of the engine, typically measured in cubic centimeters or liters.

**fuel_system:** Specifies the type of fuel system used (e.g., mpfi, 2bbl, idi).

**bore, stroke:** Represent the dimensions of the engine cylinders.

**compression_ratio:** Indicates the ratio of the volume of the cylinder when the piston is at the bottom of its stroke to the volume when the piston is at the top of its stroke.

**horsepower:** Represents the power output of the engine.

**peak_rpm:** Indicates the engine speed at which maximum horsepower is achieved.

**city_mpg, highway_mpg:** Represent the fuel efficiency of the car in city and highway driving conditions.

**price:** Represents the price of the car.

# **Exercise 5**

**Exercise 5.1 - Perform EDA on any two numerical features and find their correlation.**

```python
# Select two numerical features for EDA
feature1 = 'engine_size'
feature2 = 'horsepower'

# Calculate the correlation between the two features
correlation = df[feature1].corr(df[feature2])
print(f"Correlation between {feature1} and {feature2}: {correlation}")

# Create a scatter plot to visualize the relationship
plt.figure(figsize=(8, 6))
sns.scatterplot(x=feature1, y=feature2, data=df,hue='make', palette="viridis", s=100)
plt.title(f'Scatter Plot of {feature1} vs. {feature2}')
plt.xlabel(feature1)
plt.ylabel(feature2)
plt.show()

# Create histograms for each feature
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(df[feature1], bins=15, edgecolor='black')
plt.xlabel(feature1)
plt.ylabel('Frequency')
plt.title(f'Distribution of {feature1}')

plt.subplot(1, 2, 2)
plt.hist(df[feature2], bins=15, edgecolor='black')
plt.xlabel(feature2)
plt.ylabel('Frequency')
plt.title(f'Distribution of {feature2}')

plt.tight_layout()
plt.show()

#Boxplot
plt.figure(figsize=(20,6))
sns.boxplot(x=feature1, y=feature2, data=df)
plt.title(f"Boxplot of {feature2} vs {feature1}")
plt.show()
```

*   **Strong Positive Correlation:** engine_size and horsepower are strongly positively correlated, meaning larger engines tend to have more horsepower.
*   Engine Size Distribution: engine_size values are somewhat concentrated around a lower range with a tail extending towards larger sizes (slightly right-skewed).
*   **Horsepower Distribution:** horsepower shows a similar distribution to engine_size, slightly right-skewed with most values in a lower range and a tail towards higher horsepower.
*   **Relationship Visualization:** The scatter plot clearly demonstrates the positive trend between the two features.
*   **Potential Outliers:** There are a few potential outliers with high engine_size and horsepower values, indicated by points further away from the main cluster in the scatter plot and box plot.
*   **Boxplot Insights:** The boxplot reveals the distribution of horsepower across different ranges of engine_size. It further confirms the positive relationship and highlights potential outliers.



**Exercise 5.2 - Create a new feature 'weight_to_engine_ratio' and analyze its distribution.**

```python
# Ensure 'curb_weight' and 'engine_size' are numeric
df['curb_weight'] = pd.to_numeric(df['curb_weight'], errors='coerce')
df['engine_size'] = pd.to_numeric(df['engine_size'], errors='coerce')

# Drop rows with NaN values in these columns
df.dropna(subset=['curb_weight', 'engine_size'], inplace=True)

# Create the new feature
df['weight_to_engine_ratio'] = df['curb_weight'] / df['engine_size']

# Analyze the distribution of the new feature
plt.figure(figsize=(10, 6))
sns.histplot(df['weight_to_engine_ratio'], kde=True)
plt.title('Distribution of Weight to Engine Ratio')
plt.xlabel('Weight to Engine Ratio')
plt.ylabel('Frequency')
plt.show()

df['weight_to_engine_ratio'].describe()
```

*   Most cars have a weight-to-engine ratio between approximately 2.5 and 4. This is where the histogram's peak is located, indicating the most frequent range.
*   The distribution is slightly right-skewed. The tail of the histogram extends further to the right, implying there are some cars with higher weight-to-engine ratios than the typical range.
*   There are potential outliers. Some data points appear beyond the main body of the histogram, particularly on the higher end of the ratio. This suggests the existence of cars with unusually high weight-to-engine ratios compared to the majority.
*   There's a moderate amount of variability in the weight-to-engine ratio. The histogram is not extremely narrow or wide, suggesting a reasonable amount of variation among the cars in the dataset.
*   There are no significant gaps in the distribution. This indicates a relatively continuous range of weight-to-engine ratios, with no major categories or distinct groupings.

In the context of this automobile dataset, these observations could lead to the following inferences:

*   Most of the cars in this dataset appear to have a relatively balanced weight-to-engine ratio, likely aiming for a balance between performance and fuel efficiency.
*   The presence of some cars with higher ratios suggests potential variations in car types, such as heavier sedans, SUVs, or trucks within the dataset.
*   The outliers might represent specialized vehicles or those with specific design considerations, warranting further investigation.

By combining these observations with other analyses and domain knowledge, we can further refine our understanding of the car dataset and uncover more detailed relationships between its features.



**Exercise 5.3 - Visualize the relationship between 'drive_wheels' and 'price'.**

```python
# Convert 'price' column to numeric
df['price'] = pd.to_numeric(df['price'], errors='coerce')

# Remove rows with missing 'price' values
df.dropna(subset=['price'], inplace=True)

# Create the box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='drive_wheels', y='price', data=df)
plt.title('Relationship between Drive Wheels and Price')
plt.xlabel('Drive Wheels')
plt.ylabel('Price')
plt.show()
```

*   Price variation across drive wheels: Car prices differ based on their drive wheel type.
*   Rear-wheel drive (rwd) cars tend to be more expensive: The median price for rwd cars is higher than for fwd or 4wd.
*   Front-wheel drive (fwd) cars are generally the most affordable: They have the lowest median price and a narrower price range.
*   Four-wheel drive (4wd) cars have a wide price range: Prices for 4wd cars vary considerably, with some overlapping with rwd cars.
*   Outliers present: There are some unusually high-priced cars in each drive wheel category, especially for rwd.

In short, the plot suggests that drive wheel type is a factor influencing car price, with rear-wheel drive generally associated with higher prices.
